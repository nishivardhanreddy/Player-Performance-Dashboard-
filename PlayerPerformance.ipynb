```python
Bowlers.sort_values(by="Balls_Bowled", ascending=False)
```

```python
def balls_to_overs(balls):
    overs = balls // 6  # Full overs
    remaining_balls = balls % 6  # Remaining balls
    return f'{overs}.{remaining_balls}'
```

```python
Bowlers['Overs_Bowled'] = Bowlers['Balls_Bowled'].apply(balls_to_overs)
```

```python
Bowlers.sort_values(by="Balls_Bowled" , ascending= False)
```

```python
top_10_bowlers = Bowlers.sort_values(by='Balls_Bowled', ascending=False).head(10)
```

```python
top_players_centuries = data.groupby(["Year", "Player_Name"])["Centuries"].sum().reset_index()
top_10_century_players = top_players_centuries.groupby("Player_Name")["Centuries"].sum().nlargest(10).index
filtered_data = top_players_centuries[top_players_centuries["Player_Name"].isin(top_10_century_players)]

plt.figure(figsize=(12, 6))
sns.lineplot(data=filtered_data, x="Year", y="Centuries", hue="Player_Name", marker="o")
plt.title("Centuries Trends Over the Years")
plt.xlabel("Year")
plt.ylabel("Centuries Scored")
plt.legend(title="Top Players")
plt.show()
```

```python
data["Fours"] = pd.to_numeric(data["Fours"], errors="coerce").fillna(0).astype(int)
data["Sixes"] = pd.to_numeric(data["Sixes"], errors="coerce").fillna(0).astype(int)
data["Total_Boundaries"] = data["Fours"] + data["Sixes"]
top_boundaries = data.groupby("Player_Name")["Total_Boundaries"].sum().nlargest(10).reset_index()
```

```python
plt.figure(figsize=(12, 8))
sns.barplot(data=top_boundaries, x="Total_Boundaries", y="Player_Name", palette="viridis")
plt.title("Top Players with Most Boundaries")
plt.xlabel("Total Boundaries")
plt.ylabel("Player Name")
plt.show()
```

```python
# Calculate the average runs scored by each player
avg_runs = data.groupby("Player_Name")["Runs_Scored"].mean().reset_index()
avg_runs.columns = ["Player_Name", "Avg_Runs"]

# Sort the players by their average runs and select the top 10
top_players_avg = avg_runs.sort_values(by="Avg_Runs", ascending=False).head(10)

# Now filter the original data for only these top players
top_players_data = data[data["Player_Name"].isin(top_players_avg["Player_Name"])]

# Aggregate the total matches played and total runs scored by these top players
top_players_stats = top_players_data.groupby("Player_Name").agg(
    Total_Matches=("Matches_Batted", "sum"),
    Total_Runs=("Runs_Scored", "sum")
).reset_index()

# Create a scatter plot for total matches vs total runs
plt.figure(figsize=(10, 6))
sns.scatterplot(data=top_players_stats, x="Total_Matches", y="Total_Runs", hue="Player_Name", palette="Set2", s=100, alpha=0.7)

# Add titles and labels
plt.title("Total Matches Played vs Total Runs Scored by Top Players (Based on Avg Runs)")
plt.xlabel("Total Matches Played")
plt.ylabel("Total Runs Scored")
plt.legend(title="Player Name", bbox_to_anchor=(1.05, 1), loc='upper left')

# Display the plot
plt.show()
```

```python
# Aggregate total runs scored by each player
total_runs = data.groupby("Player_Name")["Runs_Scored"].sum().reset_index()

# Sort the players by total runs scored in descending order and select the top 10
top_10_players = total_runs.sort_values(by="Runs_Scored", ascending=False).head(10)

# Create a bar plot for total runs scored by the top 10 players
plt.figure(figsize=(12, 8))
sns.barplot(data=top_10_players, x="Player_Name", y="Runs_Scored", palette="viridis")
plt.title("Top 10 Players by Total Runs Scored in IPL")
plt.xlabel("Player Name")
plt.ylabel("Total Runs Scored")
plt.xticks(rotation=45)
plt.show()
```
